#### Ciclos de repetición (for) con if, elif y else
55. Programa que calcule y muestre el factorial de todos los números desde 1 hasta N.

```python
# ==========================================
# ACT 14
# PROBLEMA 55
# FACTORIALES DEL 1 HASTA N
# ==========================================

print("===================================")
print(" FACTORIALES DESDE 1 HASTA N ")
print("===================================")

n = int(input("Ingrese un número N: "))

if n <= 0:

    print("Debe ingresar un número mayor que cero.")

else:

    for numero in range(1, n + 1):

        factorial = 1

        for i in range(1, numero + 1):

            factorial = factorial * i

        print("Factorial de", numero, "=", factorial)

    if n == 1:

        print("\nSe calculó solamente un factorial.")

    elif n >= 2 and n <= 10:

        print("\nCantidad de factoriales calculados: ", n)

    else:

        print("\nSe calcularon muchos factoriales.")
```
